Q. 그러면 강의 중에 strides가 다르먄 말씀하신 뷰도 다른 건가요?

A. "다차원 배열의 strides만 변경 헀을 때 바라보는 data가 변경되나요?"라는 질문으로 이해해도 되나요? 만약 이 질문이라면 No입니다. 바라보는 데이터의 주소는 numpy.ndarray.data로 확인 가능합니다. 다차원 배열을 생성하고, strides를 변경해주신다음 .data로 확인해보면 동일한 곳을 가리키는것을 확인할 수 있습니다.

```python
import numpy as np

a = np.array([[1, 2, 3], 
              [4, 5, 6]])
b = a.T

# 다차원배열 a와 다차원배열 b는 다른 메모리공간에 할당되어 있습니다.
print(hex(id(a)))
print(hex(id(b)))

# 그렇지만 바라보는 데이터의 위치는 동일합니다.
print(a.data)
print(b.data)
```

Q. 딥러닝시에 마지막 layer에서 ravel과 flatten을 사용하였을 때 메모리 점유량의 차이가 있나요?

A. ravel(), flatten() 둘 다 1차원 다차원배열을 반환합니다. 두 경우 모두 반환하는 다차원배열을 저장해야 하니 메모리가 할당됩니다.

출력해보면 동일한 결과가 출력됩니다만 만들어지는 방식은 다릅니다.

ravel은 원본의 요소를 참조해서 만들어지고 flatten은 원본의 요소의 값을 복사해서 만들어집니다. 이로인한 특성은 두가지입니다.
- 값을 복사할 필요없이 참조만 하면 되니 ravel()로 반환되는 다차원 배열의 용랑이 더 작습니다.
- ravel로 만들어진 배열의 요소를 바꾸면 원본도 함께 바뀌지만 flatten으로 만들어진 배열의 요소를 바꾸더라도 복사된 값을 변경하는것이기에 원본은 아무 변화가 없습니다.

```python
import numpy as np
import sys

a = np.array([[1, 2, 3], 
              [4, 5, 6]])

ravelArray = a.ravel()
flattenArray = a.flatten()

print(ravelArray)
print(flattenArray)

print(sys.getsizeof(a))
print(sys.getsizeof(ravelArray))
print(sys.getsizeof(flattenArray))
```

정확하게 설명을 하면 strides의 숫자는 바이트를 의미합니다. strides가 (2,1)이라면, 행이 바뀔땐 2바이트, 열이 바뀔땐 1바이트 다음의 데이터를 읽어오는 것입니다. 간소화하여 이해를 돕기 위해 다차원배열의 데이터 하나가 1바이트가 되도록 일부러 int8로 선언을 했었습니다.

int8은 8비트로 정수를 표현하며 8비트 = 1바이트 입니다. 즉, int8 정수 하나는 1바이트를 차지합니다.

```python
import numpy as np

a = np.array([[1, 2, 3], 
              [4, 5, 6]], dtype='int8')
```
